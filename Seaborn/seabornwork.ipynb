WORK WITH SCATTERPLOTS!!

```python
diamonds = sns.load_dataset('diamonds')
diamonds.shape 
```

```python
diamonds.head
```

```python
diamonds = diamonds[
            diamonds.cut.isin(['Premium', 'Good', 'Very Good']) &
            diamonds.color.isin(['D','F','J'])
            ].sample(n=100, random_state=22)
diamonds.shape
```

```python
sns.set_style("darkgrid")
```

```python
sns.scatterplot(diamonds.carat, diamonds.price)
```

```python
sns.scatterplot(x='carat', y='price', data=diamonds)
```

```python
sns.scatterplot(x='x',y='y',data=diamonds)
```

these x and y values are actual variables in the data set to make a good model

```python
sns.scatterplot(x='carat', y='price', hue='cut', data=diamonds)
```

```python
sns.scatterplot(x='carat', y='price', hue='cut', data=diamonds,
                palette=['purple', 'red', '#55CCCC'])
```

```python
sns.scatterplot(x='carat', y='price', hue='depth', data=diamonds)
```

^^ looking at the depth percentage, which is a continous variable mapping the lower and higher percentages of depth

changing the plot shapes

```python
sns.scatterplot(x='carat', y='price', style='cut', data=diamonds,
                s = 100)
```

changing the sizes based off the type of cut for the diamonds

```python
sns.scatterplot(x='carat', y='price', size='cut',
                data=diamonds, 
                s=[300, 200, 100]
               )
```

sizes are more difficult to see to the human eye vs. the colors

putting it all together

```python
sns.scatterplot(x='carat', y='price', hue='cut',
                style='cut',
                data=diamonds,
                s = 150
               )
```

```python
sns.scatterplot(x='carat', y='price', hue='cut',
                style='cut',
                alpha = 0.75,
                edgecolor = 'black',
                data=diamonds,
                s = 150
               )
```

TIME FOR HEAT MAPS!! :)

```python
cars = sns.load_dataset('mpg')
```

```python
cars.head()
```

```python
cars.groupby('origin').cylinders.value_counts()
```

unstack the data

```python
(cars
 .groupby('origin')
 .cylinders
 .value_counts()
 .unstack()
 )
```

```python
origin_cyl =(cars
 .groupby('origin')
 .cylinders
 .value_counts()
 .unstack()
 .fillna(0)
 )
```

```python
sns.heatmap(origin_cyl)
```

Transpose the data so it is represented by rows of cylinders

```python
origin_cyl.T
```

```python
sns.heatmap(origin_cyl.T)
```

Correlation

```python
cars.corr()
```

```python
sns.heatmap(cars.corr())
```

values that are positively correlated are a lighter shade vice versa with the darker shade

time to choose a color palette - seaborn has over 170 different color palettes

```python
sns.heatmap(origin_cyl, cmap='Blues')
```

create a diverging palette with 0 to be the baseline

```python
sns.heatmap(cars.corr(), cmap='RdBu', center=0)
```

annotating the heatmap 

```python
sns.heatmap(origin_cyl, cmap='Blues', annot=True)
```

remove the significant digits

```python
sns.heatmap(origin_cyl, cmap='Blues', annot=True, fmt=".0f") #.0 means there are zero decimals (if you wanted 1 decimal it would be .1, 2 decimals .2, etc.
                                                             #f represents a floating point number
```

```python
sns.heatmap(origin_cyl, cmap='Blues', annot=True, fmt=".0f",
            annot_kws={
                'fontsize': 16,
                'fontweight': 'bold',
                'fontfamily': 'garamond'
            })
```

But what if I want spaces between my heatmap??

```python
sns.heatmap(cars.corr(), cmap='RdBu', center=0, linewidth=1)
```

create your own labels

```python
labels = ['MPG', 'Cylinders', 'Displacement', 'Horsepower',
          'Weight', 'Acceleration', 'Model Year']
```

```python
sns.heatmap(cars.corr(), 
            cmap='RdBu',
            xticklabels=labels,
            yticklabels=labels
           )
```

Remove labels on either axis

```python
sns.heatmap(cars.corr(), 
            cmap='RdBu', 
            xticklabels=False,
            yticklabels=labels
           );
```

Creating a perfect square on the heatmap

```python
sns.heatmap(cars.corr(), 
            cmap='RdBu',
            square=True
           )
```
